#### 1. Run SQL commands using PySpark.
#### 2. Spark dataframe reader API.

### 1.Run SQL commands using PySpark.

```python
spark.sql('SELECT * FROM json.`/Volumes/gizmobox/landing/operations_volume/customers/`')
```

```python
customers_df = spark.sql('SELECT * FROM json.`/Volumes/gizmobox/landing/operations_volume/customers/`')
display(customers_df)
```

```python
temp_customers_df = spark.sql('''CREATE OR REPLACE TEMP VIEW v_customer 
                         AS
                         SELECT * FROM json.`/Volumes/gizmobox/landing/operations_volume/customers/`''')
```

```python
%sql
SELECT * FROM v_customer
```

### 2. Spark dataframe reader API.

```python
df_customers = spark.read.format('json').load('/Volumes/gizmobox/landing/operations_volume/customers/')
display(df_customers)
```
