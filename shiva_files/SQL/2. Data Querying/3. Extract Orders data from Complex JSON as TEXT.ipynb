## Extract data from ORDERS JSON file
### 1. Query orders file from `JSON` format.
### 2. Query orders file from `TEXT` format. 
### 3. Create orders view in bronze schema.

### 1. Query orders file from `JSON` format.

```sql
SELECT * FROM json.`/Volumes/gizmobox/landing/operations_volume/orders/`;
```

- ##### There are some corrupt records so we are getting error. Eg: Records 2, value for `order_date` is not enclosed in double quotes. 

- ##### Corrupt records are shown under `_corrupt_records` column.

- ##### So using `text` format to query records as `text` format will give the output as it is in the source data file in a single `value` column.



### 2. Query orders file from `TEXT` format.

```sql
SELECT * FROM text.`/Volumes/gizmobox/landing/operations_volume/orders/`;
```

### 3. Create orders view in bronze schema.

```sql
CREATE OR REPLACE VIEW gizmobox.bronze.v_orders
AS
SELECT * FROM text.`/Volumes/gizmobox/landing/operations_volume/orders/`;
```

```sql
SELECT * FROM gizmobox.bronze.v_orders;
```
