### Transform Addresses Data
#### 1. Create one record for each customer with 2 sets of address column, 1 for shipping and 1 for billing address.

#### 2. Write transformed data to Silver path.

#### 1. Create one record for each customer with 2 sets of address column, 1 for shipping and 1 for billing address.

```sql
SELECT * FROM gizmobox.bronze.v_addresses;
```

```sql
SELECT * FROM (
SELECT
      customer_id,
      address_type,
      address_line_1,
      city,
      state,
      postcode
FROM
      gizmobox.bronze.v_addresses)
PIVOT (max(address_line_1) AS address_line_1,
       max(city) AS city,
       max(state) AS state,
       max(postcode) AS postcode
       FOR address_type IN ('shipping', 'billing')
);
```

```sql
use catalog gizmobox;
use schema silver;
```

```sql
CREATE OR REPLACE TABLE gizmobox.silver.addresses
AS
SELECT * FROM (
SELECT
      customer_id,
      address_type,
      address_line_1,
      city,
      state,
      postcode
FROM
      gizmobox.bronze.v_addresses)
PIVOT (max(address_line_1) AS address_line_1,
       max(city) AS city,
       max(state) AS state,
       max(postcode) AS postcode
       FOR address_type IN ('shipping', 'billing')
);
```

```sql
SELECT * FROM gizmobox.silver.addresses;
```
