### For each customer, produce the following monthly summary per month.
#### 1. Total_orders.
#### 2. Total_items_bought.
#### 3. Total_amount_spent. 


```sql
select * from gizmobox.silver.orders;
```

```sql
SELECT
customer_id,
date_format(order_date, 'yyyy-MM') as month,
count(DISTINCT order_id) AS total_orders_per_month
FROM gizmobox.silver.orders GROUP BY customer_id, month order by customer_id;
```

```sql
SELECT 
customer_id,
date_format(order_date, 'yyyy-MM') as month,
sum(quantity) AS total_items_bought
FROM gizmobox.silver.orders WHERE order_status IN ('Completed', 'Shipped') GROUP BY customer_id, month ORDER BY customer_id; 
```

```sql
SELECT 
  customer_id,
  date_format(order_date, 'yyyy-MM') as month,
  sum(price) as total_amount_spent_per_month
FROM gizmobox.silver.orders GROUP BY customer_id, month ORDER BY customer_id;
```

```sql
SELECT 
  customer_id,
  date_format(order_date, 'yyyy-MM') as month,
  count(DISTINCT order_id) AS total_orders_per_month,
  sum(quantity) AS total_items_bought,
  sum(price*quantity) as total_amount_spent_per_month
FROM gizmobox.silver.orders GROUP BY customer_id, month ORDER BY customer_id;
```

```sql
CREATE OR REPLACE TABLE gizmobox.gold.ORDERS_SUMMARY_MONTHLY
SELECT 
  customer_id,
  date_format(order_date, 'yyyy-MM') as month,
  count(DISTINCT order_id) AS total_orders_per_month,
  sum(quantity) AS total_items_bought,
  sum(price*quantity) as total_amount_spent_per_month
FROM gizmobox.silver.orders GROUP BY customer_id, month ORDER BY customer_id;
```

```sql
SELECT * FROM gizmobox.gold.ORDERS_SUMMARY_MONTHLY;
```
