### For each customer, produce the following monthly summary per month.
#### 1. Total_orders.
#### 2. Total_items_bought.
#### 3. Total_amount_spent. 


```python
orders = spark.sql('select * from gizmobox.silver.py_orders_final')
display(orders)
```

```python
from pyspark.sql import functions as F

total_orders = (
                  orders
                     .withColumn('order_month', F.date_format('transaction_timestamp', 'yyyy-MM'))
                     .groupBy('customer_id', 'order_month')
                     .agg(
                        F.countDistinct('order_id').alias('total_orders'),
                        F.sum('quantity').alias('total_items_bought'),
                        F.sum(F.col('price') * F.col('quantity')).alias('total_amount') 
                     )
               )
display(total_orders)
```

```python
total_orders.writeTo('gizmobox.gold.py_orders_monthly_summary').createOrReplace()
```
