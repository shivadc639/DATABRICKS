### Extract data from Membership IMAGE files.
#### 1. Query Membership file using binaryFile format.
#### 2. Create Membership table in Bronze schema.

```python
%fs ls '/Volumes/gizmobox/landing/operations_volume/memberships/'
```

#### 1. Query Membership file using binaryFile format.

```python
df_membership = spark.read.format("binaryFile").load('/Volumes/gizmobox/landing/operations_volume/memberships/*')
display(df_membership)
```

####  2. Create Membership table in Bronze schema.

```python
df_membership.writeTo("gizmobox.bronze.py_memberships").createOrReplace()
```

```python
%sql
SELECT * FROM gizmobox.bronze.py_memberships;
```
