### Extract data from Orders JSON file.
#### 1. Query Orders file using JSON format.
#### 2. Query Orders file using TEXT format.
#### 3. Create Orders table in Bronze schema.

```python
%fs ls '/Volumes/gizmobox/landing/operations_volume/orders/'
```

#### 1. Query Orders file using JSON format.

```python
df_orders_json = spark.read.json('/Volumes/gizmobox/landing/operations_volume/orders/')
display(df_orders_json)
```

####  2. Query Orders file using TEXT format.

```python
df_orders_text = spark.read.text('/Volumes/gizmobox/landing/operations_volume/orders/')
display(df_orders_text)
```

#### 3. Create Orders table in Bronze schema.

```python
df_orders_text.writeTo("gizmobox.bronze.py_orders").createOrReplace()
```

```python
%sql
SELECT * FROM gizmobox.bronze.py_orders;
```
