#### 1. Extract DATE and TIME from payment_timestamp and create new columns payment_date and payment_time.

```sql
SELECT * FROM gizmobox.bronze.payments;
```

```sql
SELECT *, TO_DATE(payment_timestamp) AS payment_date, date_format(payment_timestamp, 'HH:mm:ss') AS payment_time FROM gizmobox.bronze.payments;
```

### 2. Map payment status to descriptive values
#### (1 -> Success, 2 -> Pending, 3 -> Cancelled, 4 -> Failed).

```sql
SELECT payment_id, order_id, payment_timestamp, 
      CASE WHEN  psyment_status = 1 THEN 'Success'
           WHEN psyment_status =  2 THEN 'Pending'
           WHEN psyment_status = 3 THEN 'Cancelled'
           ELSE 'Failed'
      END AS payment_status,
      payment_method,
      TO_DATE(payment_timestamp) AS payment_date, 
      date_format(payment_timestamp, 'HH:mm:ss') AS payment_time
FROM gizmobox.bronze.payments;
```

### 3. Write transformed data to Silver schema.

```sql
CREATE OR REPLACE TABLE gizmobox.silver.payment
AS
SELECT payment_id, order_id, 
      CASE WHEN  psyment_status = 1 THEN 'Success'
           WHEN psyment_status =  2 THEN 'Pending'
           WHEN psyment_status = 3 THEN 'Cancelled'
           ELSE 'Failed'
      END AS payment_status,
      payment_method,
      TO_DATE(payment_timestamp) AS payment_date, 
      date_format(payment_timestamp, 'HH:mm:ss') AS payment_time
FROM gizmobox.bronze.payments;
```
