### Transform Membership Data
#### 1. Extract customer_id from the file path.

#### 2. Write transformed data to Silver path.

#### 1. Extract customer_id from the file path.

```sql
SELECT * FROM gizmobox.bronze.v_memberships;
```

```sql
SELECT
      path,
      split_part(split_part(path, '/',8),'.', 1) AS CUSTOMER_ID,
      modificationTime,
      length,
      content
FROM
      gizmobox.bronze.v_memberships;
```

```sql
use catalog gizmobox;
use schema silver;
```

```sql
CREATE OR REPLACE TABLE gizmobox.silver.memberships
AS
SELECT
      split_part(split_part(path, '/',8),'.', 1) AS CUSTOMER_ID,
      content AS MEMBERSHIP_CARD
FROM
      gizmobox.bronze.v_memberships;
```

```sql
SELECT * FROM gizmobox.silver.memberships;
```

```sql
DESCRIBE EXTENDED gizmobox.silver.memberships;
```
