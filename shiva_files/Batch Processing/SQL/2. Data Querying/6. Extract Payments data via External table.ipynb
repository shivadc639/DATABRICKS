## Extract data from Payments csv file
#### 1. List the files in the payments folder.
#### 2. Create External table.
#### 3. Effect of Adding/Updating/Deleting files.
#### 4. Effect of dropping the table

- ##### We dont have Payments folder/files in the Unity Catalog volume.

- ##### So to read files, creating External table over the payments files which are available in blob.




### 1.List the files in the payments folder.

```sql
%fs ls 'abfss://dbdeblob@dbdeassociatesa.dfs.core.windows.net/landing/external/payments/'
```

### 2. Create External table.

```sql
CREATE TABLE IF NOT EXISTS gizmobox.bronze.payments
      (payment_id integer, order_id integer, payment_timestamp TIMESTAMP, psyment_status INTEGER, payment_method STRING)
USING CSV
OPTIONS
      (delimiter = ",",
      header = 'TRUE')
LOCATION 'abfss://dbdeblob@dbdeassociatesa.dfs.core.windows.net/landing/external/payments/'
```

```sql
SELECT * FROM gizmobox.bronze.payments;
```

```sql
DESCRIBE EXTENDED gizmobox.bronze.payments;
```

### 3. Effect of Adding/Updating/Deleting files.

- ##### Added `payments_2025_02.csv` file to blob and running `SELECT * FROM gizmobox.bronze.payments` will result in still 87 rows. Thus `REFRESH TABLE` command should be used when any file is deleted/updated/inserted.



```sql
REFRESH TABLE gizmobox.bronze.payments;
```

```sql
SELECT * FROM gizmobox.bronze.payments;
```

- ##### After `REFRESH TABLE` results 114 rows including records of newly added `payments_2025_02.csv` file.

### 4. Effect of dropping the table

```sql
DROP TABLE IF EXISTS gizmobox.bronze.payments;
```

- ##### Dropping table will drop the metadata of the file which is created and will not drop the file itself.
