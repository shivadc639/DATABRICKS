## Extract data from Refunds table in SQL server DB.
#### 1. Create Bronze schema in hive metastore.
#### 2. Create External table using `JDBC`.

- ##### Refunds is an table in SQL Server database. We are connecting to SQL Server database to read Refunds table.

- ##### External tables can only be created via JDBC in hive metastore schema. If you are working with Unity Catalog, you can use Lakehouse Federation to access tables directly. 



### 1.Create Bronze schema in hive metastore.

```sql
CREATE SCHEMA IF NOT EXISTS hive_metastore.bronze;
```

### 2. Create External table using `JDBC`.

```sql
use catalog gizmobox;
use schema landing;

CREATE OR REPLACE GLOBAL TEMPORARY VIEW refunds
USING JDBC
OPTIONS (
      url 'jdbc:sqlserver://gizmobox-serverde.database.windows.net:1433;databaseName=gizmosqldb;encrypt=true;trustServerCertificate=false;loginTimeout=30;',
      dbtable 'Refunds',
      user 'gizmoboxadmin',
      password 'Selfconfidence45'
      );
```

```sql
SELECT * FROM refunds;
```

```sql
DESC EXTENDED hive_metastore.bronze.refunds;
```
