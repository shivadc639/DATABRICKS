### Extract data from Customer JSON file.
#### 1. Query single file.
#### 2. Query list of files using wildcard.
#### 3. Query all files in the folder.
#### 4. Select file Metadata.
#### 5. Create table in Bronze schema.

#### 1. Query single file.

```python
%fs ls '/Volumes/gizmobox/landing/operations_volume/customers/'
```

```python
df_customers = spark.read.format("json").load("/Volumes/gizmobox/landing/operations_volume/customers/customers_2024_10.json")
display(df_customers)
```

####  2. Query list of files using wildcard.

```python
df_customers_wildcard = spark.read.format("json").load('/Volumes/gizmobox/landing/operations_volume/customers/customers_2024_*.json')
display(df_customers_wildcard)
```

#### 3. Query all files in the folder.

```python
df_customers_all = spark.read.format("json").load('/Volumes/gizmobox/landing/operations_volume/customers/')
display(df_customers_all)
```

#### 4. Select file Metadata. 

```python
df_customers_with_metadata = df_customers_all.select("_metadata.file_name", "*")
display(df_customers_with_metadata)
```

#### 5. Create table in Bronze schema.

```python
df_customers_with_metadata.write.format("delta").mode("overwrite").saveAsTable("gizmobox.bronze.py_customers")
```
