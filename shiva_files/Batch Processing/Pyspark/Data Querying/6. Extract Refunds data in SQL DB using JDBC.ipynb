## Extract data from Refunds table in SQL server DB.
#### 1. Read Refunds data via JDBC.
#### 2. Write Refunds data in Bronze schema.

- ##### Refunds is an table in SQL Server database. We are connecting to SQL Server database to read Refunds table.




### 1.Read Refunds data via JDBC.

```python
df_refunds = (
              spark.read.format('jdbc')
                        .option('url', 'jdbc:sqlserver://gizmobox-serverde.database.windows.net:1433;databaseName=gizmosqldb')
                        .option('dbtable', 'Refunds')
                        .option('user', 'gizmoboxadmin')
                        .option('password', 'Selfconfidence45')
                        .load()
)
display(df_refunds)
```

### 2. Write Refunds data in Bronze schema.

```python
df_refunds.writeTo("gizmobox.bronze.py_refunds").createOrReplace()
```

```python
%sql
SELECT * FROM gizmobox.bronze.py_refunds;
```
