#### 1. Read all the addresses file

```python
df_payments = (spark.read
                .format("csv").option('delimiter', ",")
                .option('header', "False")
                .load("abfss://dbdeblob@dbdeassociatesa.dfs.core.windows.net/landing/external/payments/")
                )
display(df_payments)
```

```python
sql_payments_schema = 'payment_id integer, order_id integer, payment_timestamp TIMESTAMP, psyment_status INTEGER, payment_method STRING'
```

```python
df_payments = (spark.read
                .format("csv").option('delimiter', ",")
                .schema(sql_payments_schema)
                .load("abfss://dbdeblob@dbdeassociatesa.dfs.core.windows.net/landing/external/payments/")
                )
display(df_payments)
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, TimestampType, StringType

py_payments_schema = StructType(
    [
        StructField('payment_id', IntegerType()),
        StructField('order_id', IntegerType()),
        StructField('payment_timestamp', TimestampType()),
        StructField('payment_status', IntegerType()),
        StructField('payment_method', StringType())
    ]
)
```

```python
df_payments = (spark.read
                .format("csv").option('delimiter', ",")
                .schema(py_payments_schema)
                .load("abfss://dbdeblob@dbdeassociatesa.dfs.core.windows.net/landing/external/payments/")
                )
display(df_payments)
```

#### 2. Create Payments table in Bronze schema.

```python
df_payments.writeTo("gizmobox.bronze.py_payments").createOrReplace()
```
