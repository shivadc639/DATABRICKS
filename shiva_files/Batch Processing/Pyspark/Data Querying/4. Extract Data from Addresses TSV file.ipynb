### Extract data from Addresses TSV files.
#### 1. Read all the Addresses file
#### 2. Create Addresses table in Bronze schema.

```python
%fs ls '/Volumes/gizmobox/landing/operations_volume/addresses/'
```

#### 1. Read all the addresses file

```python
df_addresses = (spark.read
                .format("csv").option('delimiter', "\t")
                .option('header', "TRUE")
                .load("/Volumes/gizmobox/landing/operations_volume/addresses/")
                )
display(df_addresses)
```

#### 2. Create Addresses table in Bronze schema.

```python
df_addresses.writeTo("gizmobox.bronze.py_addresses").createOrReplace()
```

```python
%sql
SELECT * FROM gizmobox.bronze.py_addresses;
```
