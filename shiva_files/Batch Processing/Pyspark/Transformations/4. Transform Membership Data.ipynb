### Transform Membership Data
#### 1. Extract customer_id from the file path.

#### 2. Write transformed data to Silver path.

#### 1. Extract customer_id from the file path.

```python
df_membership_raw = spark.sql('SELECT * FROM gizmobox.bronze.v_memberships')
display(df_membership_raw)
```

```python
from pyspark.sql import functions as F
df_membership_transformed = (df_membership_raw.select(
                                        F.regexp_extract('path', r".*/([0-9]+)\.png$", 1).alias("customer_id"),
                                        'content'
                            )
)
display(df_membership_transformed)
```

```python
df_membership_transformed.writeTo('gizmobox.silver.py_memberships').createOrReplace()
```
