### Transform Addresses Data
#### 1. Create one record for each customer with 2 sets of address column, 1 for shipping and 1 for billing address.

#### 2. Write transformed data to Silver path.

#### 1. Create one record for each customer with 2 sets of address column, 1 for shipping and 1 for billing address.

```python
df_address_raw = spark.sql('SELECT * FROM gizmobox.bronze.v_addresses');
display(df_address_raw)
```

```python
from pyspark.sql import functions as F

df_address_transformed = (df_address_raw
                            .groupBy('customer_id')
                            .pivot('address_type', ['billing', 'shipping'])
                            .agg(
                                 F.max(F.col('address_line_1')).alias('address_line_1'),
                                 F.max(F.col('city')).alias('city'),
                                 F.max(F.col('state')).alias('state'),
                                 F.max(F.col('postcode')).alias('postcode')
                                 )
                         )
display(df_address_transformed)
```

```python
df_address_transformed.writeTo('gizmobox.silver.py_address').createOrReplace()
```
