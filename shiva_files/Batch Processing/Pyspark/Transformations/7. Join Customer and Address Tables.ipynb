### Join Customer and Address table to create customer_address which contains the address of each customer on the same record.

```python
df_customers = spark.sql('SELECT * FROM gizmobox.silver.customer');
display(df_customers)
```

```python
df_addresses = spark.sql('SELECT * FROM gizmobox.silver.addresses;')
display(df_addresses)
```

```python
df_joined  = (df_customers
                        .join(df_addresses, 
                              df_customers.customer_id == df_addresses.customer_id, 
                              'inner'
                              )
            )
display(df_joined)
```

```python
df_customers_addresses = (df_joined 
                          .select(
                                    df_customers.customer_id.alias('customer_id'),
                                    df_customers.customer_name.alias('customer_name'),
                                    df_customers.email,
                                    df_customers.date_of_birth,
                                    df_customers.member_since,
                                    df_customers.telephone,
                                    df_addresses.shipping_address_line_1,
                                    df_addresses.shipping_city,
                                    df_addresses.shipping_state,
                                    df_addresses.shipping_postcode,
                                    df_addresses.billing_address_line_1,
                                    df_addresses.billing_city,
                                    df_addresses.billing_state,
                                    df_addresses.billing_postcode
                                )       
)
display(df_customers_addresses)
```

```python
df_customers_addresses.writeTo('gizmobox.gold.py_customers_addresses').createOrReplace()
```
